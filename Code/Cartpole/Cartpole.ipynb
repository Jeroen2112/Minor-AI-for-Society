```python
import gym
import os
from stable_baselines3 import A2C, PPO, DQN
from stable_baselines3.common.env_util import make_vec_env
```

```python
env = make_vec_env("CartPole-v1", n_envs=4)
```

```python
log_path = os.path.join('Training', 'Logs')
a2c_path = os.path.join('Training', 'Saved Models', 'A2C_Cartpole')
dqn_path = os.path.join('Training', 'Saved Models', 'DQN_Cartpole')
ppo_path = os.path.join('Training', 'Saved Models', 'PPO_Cartpole')
```

```python
model = A2C("MlpPolicy", env, verbose=1, tensorboard_log=log_path, device='cuda')
model.learn(total_timesteps=200000, log_interval=100)
model.save(a2c_path)
```

```python
del model
```

```python
model = DQN("MlpPolicy", env, verbose=1, tensorboard_log=log_path, device='cuda')
model.learn(total_timesteps=200000, log_interval=100)
model.save(dqn_path)
```

```python
del model
```

```python
model = PPO("MlpPolicy", env, verbose=1, tensorboard_log=log_path, device='cuda')
model.learn(total_timesteps=200000, log_interval=100)
model.save(ppo_path)
```

```python
del model
```
